```python
time = [4861.38,4238.74,4211.82,4197.73,4219.55]
log_time = []
log_p = []
p = [1,2,4,10,20]
speed_up = []
efficiency = []
```

```python
for i in range(5):
    log_p.append(round(math.log(p[i]),2))
    speed_up.append(round((time[0] / time[i]),2))
    efficiency.append(round((speed_up[i]/p[i]),2))
```
